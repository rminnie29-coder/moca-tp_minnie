```python
import urllib.request as req
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.moca.taipei/tw/ExhibitionAndEvent/Events"
resp = req.urlopen(url)
content = resp.read()
html = BeautifulSoup(content, "html.parser")

rs = html.find_all("a", {"class": "textFrame cg cgB"})

table = []

for r in rs:
    # 連結
    href = r.get("href")
    if href and not href.startswith("http"):
        href = "https://www.moca.taipei" + href

    # 展覽名稱與副標題
    title = r.find("h3", {"class": "imgTitle"})
    subtitle = r.find("h4", {"class": "imgSubTitle"})
    title_text = title.get_text(strip=True) if title else None
    subtitle_text = subtitle.get_text(strip=True) if subtitle else None

    # 日期、時間區塊
    day = r.find_all("span", {"class": "day"})
    en_day = r.find_all("span", {"class": "en"})
    time = r.find_all("p", {"class": "text"})

    if len(day) >= 2:
        start_day = day[0].get_text(strip=True)
        end_day = day[1].get_text(strip=True)
    elif len(day) == 1:
        start_day = day[0].get_text(strip=True)
        end_day = time
    else:
        start_day = end_day = None

    # 英文星期（如 Mon.、Thu.）
    if len(en_day) >= 2:
        start_en = en_day[0].get_text(strip=True)
        end_en = en_day[1].get_text(strip=True)
    elif len(en_day) == 1:
        start_en = en_day[0].get_text(strip=True)
        end_en = None
    else:
        start_en = end_en = None

    # 加入表格
    row = {
        "名稱": title_text,
        "副標題": subtitle_text,
        "開始時間": start_day,
        "開始星期": start_en,
        "結束時間": end_day,
        "結束星期": end_en,
        "連結": href
    }
    table.append(row)

df = pd.DataFrame(table)
df.to_csv("tabelog.csv", encoding="utf-8")
df
```
